```python
#Read the companies file from csv 
r = pd.read_csv("companies.csv")
r.head() #not necessary, but to check if data is read correctly 
```

```python
#1.
#resd companies csv file into dataframe
companiesdf = pd.DataFrame(r)
companiesdf.head() #not necessary, but to check if data is read correctly
```

```python
#print a list of companies based in the USA, and name it as "usadf"
usadf = companiesdf[companiesdf["country"] == "USA"]
print(usadf)
#or 
usadf.head() #not necessary, but to have a better look on how usadf looks
```

```python
#2.
#Sort (in Python) the companies in descending order of revenue
compdfsort = companiesdf.sort_values(by = "annualRevenue", ascending= False)
#generate a bar chart 
compdfsort.plot.bar(x = "companyName", y = "annualRevenue")
```

```python
#3.
#Generate a bubble chart showing companies by the number of employees
import matplotlib.pyplot as plt
comsort2 = companiesdf.sort_values(by = "employees", ascending= False)
comsort2.plot.scatter(x = "companyName", y = "employees")
#Interpretation: 
#The number of employees is arranged in desceding order. From the chart, we can see that Amazon has the most employees among the firms. 
#The number of employees in Amazon is around 4 times more than the firm with the second most employee. 
#Numbers of employees come from the top Amazon, Daimler AG, Samsung Electronics, Microsoft, Apple, Google, Alibaba, then to the least Sony.
```

```python
#4.
#Generate a scatterplot of revenue
import seaborn as sns
lm = sns.lmplot(x="annualRevenue", y="employees",data=companiesdf, line_kws={'color': 'red'}) 

#subtitle -> suptitle
fig = lm.fig
fig.suptitle ("Correlation-Revenue&Employees")

#Brief interpretation: From the plot we can see that generally Annual Revenue and Employees follow linear model and seems to be positively related.
#But because it is a small dataset, we cannot conclude that two variables (Annual Revenue and number of Employees) are positively correlated.
```

```python
#5.
#Generate two pie charts of companies by revenue
companyName = companiesdf["companyName"]
companiesdf.plot.pie(x="companyName", y="annualRevenue", labels=companyName, title="Companies by revenue", figsize=(10,10))
plt.show()
#by the number of employees
companiesdf.plot.pie(x="companyName", y="employees", labels=companyName, title="Companies by employee", figsize=(10, 10))
```
